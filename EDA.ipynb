```python
plt.figure(figsize=(10, 6))
sns.countplot(data=train_data, x='Choice', hue='Choice')
plt.xlabel('Choice')
plt.ylabel('Count')
plt.legend(title='Choice')
plt.show()
```

```python
sns.pairplot(train_data)
```

```python
columns = train_data.columns
columns = columns.drop('Choice')
```

```python
n_rows = 2
n_cols = 11

plt.figure(figsize=(20,10))
for index in range(22):
    plt.subplot(n_rows, n_cols, index+1)
    sns.boxplot(x = 'Choice', y = columns[index], data = train_data)
    plt.xlabel('Choice')
    plt.ylabel(columns[index])

plt.tight_layout()
plt.show()
```

```python
n_rows = 2
n_cols = 11

plt.figure(figsize=(20,10))
for index in range(22):
    plt.subplot(n_rows, n_cols, index+1)
    sns.violinplot(x = 'Choice', y = columns[index], data = train_data)
    plt.xlabel('Choice')
    plt.ylabel(columns[index])

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(20,15))
sns.heatmap(train_data.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.show()
```
